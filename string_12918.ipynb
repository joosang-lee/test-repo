```python
def solution(s):  
  if (len(s) == 4)|(len(s) == 6):
    result = []
    for syllable in s:
      if syllable in [str(i) for i in range(10)]:
        result.append(True)
      else:
        result.append(False)
    if False in result:
      return False
    else:
      return True
  else:
    return False
```
